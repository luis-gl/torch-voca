```python
import torch
import seaborn as sns
from scipy.stats import norm
from utils.init_trunc_norm import truncated_normal_
```

#### Según la documentación de Tensorflow: https://www.tensorflow.org/api_docs/python/tf/compat/v1/truncated_normal_initializer, la inicialización de parámetros en un rango de 2 desviaciones estándar.

```python
size = (100,100)
std = 0.02
tensor = torch.zeros(size)
tensor = truncated_normal_(tensor, std=std)

print(tensor.numpy().shape)
print(bool(torch.sum(tensor >= -2.0*std).item()))
print(bool(torch.sum(tensor <= 2.0*std).item()))
```

```python
sns.distplot(tensor.numpy(), fit=norm)
```
